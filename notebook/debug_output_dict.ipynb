```python
# debuging by creating another model
from keras.layers import Input, LSTM, Dense, Embedding, Bidirectional, Flatten, Reshape, Lambda, TimeDistributed, concatenate, multiply
from keras import optimizers
from keras import backend as K
from keras.models import Model
```

```python
seqlen = 5
storylen= 4
vocab_size = 8
batchsize = 2
output_dim = 50

x = np.array([[[np.random.randint(vocab_size) for _ in range(seqlen)] for _ in range(storylen)] for _ in range(batchsize)])
print("x: \n", x)
pretrained_embeddings = np.array([[i for _ in range(output_dim)] for i in range(vocab_size)])
print("pretrained_embeddings: \n", pretrained_embeddings)

inputs = Input(shape=(storylen, seqlen))
embed_layer = TimeDistributed(Embedding(input_dim = vocab_size, output_dim=output_dim,
                          input_length=seqlen, mask_zero = True,
                          weights=[pretrained_embeddings], trainable=False))
embeddings = embed_layer(inputs)
model = Model(inputs=[inputs], outputs=embeddings)
# no training
model.compile(optimizer='rmsprop', loss='binary_crossentropy')
print("embeddings: \n",  model.predict(x))
```
